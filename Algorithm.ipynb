```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Load the dataset
df = pd.read_csv('countries.csv', encoding='iso-8859-1')
```

```python
# Check the first few rows of the dataset
print(df.head())

# Check the shape of the dataset
print(df.shape)

# Check the data types of the columns
print(df.dtypes)

# Check the summary statistics of the dataset
print("Summary:\n",df.describe(),"\n")


# Check for missing values in the dataset
print("missing:\n",df.isnull().sum())

#Find the duplicates

print("duplicates:\n",df.duplicated().sum())
```

```python
#Replace null values
df = df.fillna(df.mean())

#df.replace(np.nan,'0',inplace = True)

#Check the changes now
df.isnull().sum()
```

```python
plt.figure(figsize=(60, 15))
sns.barplot(x='Country', y='Population (millions)', data=df)
plt.xticks(rotation=45, fontsize=25)
plt.yticks(fontsize=20)
plt.xlabel('Country', fontsize=16)
plt.ylabel('Population (millions)', fontsize=20)
plt.title('Population by Country', fontsize=20)
plt.show()
```

```python
df['Carbon Footprint'] = pd.to_numeric(df['Carbon Footprint'], errors='coerce')
df.dropna(subset=['Carbon Footprint'], inplace=True)
#Boxplot

df[['Carbon Footprint']].boxplot()
```

```python
#Correlation 

df.corr()
#Correlation plot

sns.heatmap(df.corr())
```

```python
# Remplacement des valeurs aberrantes par la moyenne
mean = df['Carbon Footprint'].mean()
df.loc[(df['Carbon Footprint'] < 0) | (df['Carbon Footprint'] > 5), 'Carbon Footprint'] = mean
```

```python
df['Carbon Footprint'] = df['Carbon Footprint'].interpolate(method='linear')
```
